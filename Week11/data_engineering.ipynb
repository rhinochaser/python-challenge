```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# files to load
hawaii_measure_load = "Resources/data/hawaii_measurements.csv"
hawaii_station_load = "Resources/data/hawaii_stations.csv"

# read files
measure_data = pd.read_csv(hawaii_measure_load)
station_data = pd.read_csv(hawaii_station_load)

# combine the files
station_measure_data = pd.merge(station_data, measure_data, how= 'left', on=['station', 'station'])

# display the files
station_measure_data.head()
```

```python
# Change NaNs to 0.0
station_measure_data = station_measure_data.replace(np.nan, 0)
# verify NaNs to 0.0
station_measure_data.head()
```

```python
# export cleaned csv file
station_measure_data.to_csv('Resources/data/clean_station_measure_data.csv', index = False, header = True)
```
